```python
def hotel_cost(nights):
    cost_per_night=140
    total_hotel_cost=nights*cost_per_night
    return total_hotel_cost
    
def plane_ride_cost(city):
    if city=="Charlotte":
        plane_total_cost=183
    elif city=="Tampa":
        plane_total_cost=220
    elif city=="Pittsburgh":
        plane_total_cost=222
    else:
        plane_total_cost=475
    return plane_total_cost

def rental_car_cost(days):
    daily_rate = 40
    base_cost = days * daily_rate
    discount = 0

    if days >= 7:
        discount = 50
    elif days >= 3:
        discount = 20

    final_car_cost = base_cost - discount
    return final_car_cost

def trip_cost(city,days,spending_money):
    return hotel_cost(days) + plane_ride_cost(city) + rental_car_cost(days) + spending_money

print("----TRIP COST CALCULATOR----")
nights=int(input("Enter the number of nights for stay:"))
hotel_cost(nights)
city=input("Enter the city:")
plane_ride_cost(city.title())
days=int(input("Enter the number of days:"))
rental_car_cost(days)
spending_money=int(input("Enter extra spending money:"))
print("Total trip cost:")
trip_cost(city,days,spending_money)

    
```
